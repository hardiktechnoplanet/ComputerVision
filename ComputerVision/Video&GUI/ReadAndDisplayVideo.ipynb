```python
# Import module
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the video using video reader object
cap=cv2.VideoCapture('chaplin.mp4')

# Check if video is loaded successfully 
if(cap.isOpened()==False):
    print("Error opening video file")
```

```python
# Read Frames
# ret is the returned value, frame is the returned frame
ret,frame=cap.read()
```

```python
# Display frame
# Display using Matplotlib
plt.imshow(frame[:,:,::-1])
```

```python
# Display video
# Read until video is complete
while(cap.isOpened()):
    # Capture frame by frame
    ret,frame=cap.read()
    
    if(ret==True):
        cv2.imshow("video",frame)
        # Wait for 25 ms before moving to the next frame
        cv2.waitKey(25)& 0xFF #Adding the & 0xFF returns only the last 8 bit of the number. 
        #This is useful because without it the pressed key may return different values while NumLock is turned on or off
    else:
        # Break the loop
        break
cv2.destroyWindow("video")
```
