```python
# Import module
import cv2
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
# Create a video capture object "cap"
cap=cv2.VideoCapture('chaplin.mp4')
```

# Get video properties 

```python
# Get the frame width
width =cap.get(3)

# Get frame height
height=cap.get(4)
print(width,height)
```

```python
# Read the first frame
ret,frame=cap.read()
```

```python
matplotlib.rcParams['figure.figsize'] = (10.0, 10.0)
plt.imshow(frame[:,:,::-1])
```

# Set Video Properties

```python
# Width
ret = cap.set(3, 320)
print(ret)
# Height
ret = cap.set(4, 180)
print(ret)
```

False means, we can not change these default properties

```python
cap.release()
```
