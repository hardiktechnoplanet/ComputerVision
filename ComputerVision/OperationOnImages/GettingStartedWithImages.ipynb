```python
#Import libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (6.0, 6.0)
matplotlib.rcParams['image.cmap'] = 'gray'
```

```python
#Load and read the image
imagePath="number_zero.jpg"

#Read the image in Grayscale format
#Second parameter in below funtion can be 0,1,-1
#0=Grayscale, 1=Color,-1=includes Alpha channel
testImage=cv2.imread(imagePath,0)
print(testImage)
```

```python
#Image properties
print("Data Type ={}\n".format(testImage.dtype))
print("Object Type ={}\n".format(type(testImage)))
print("Image Dimensions ={}\n".format(testImage.shape))
```

Manipulating Group of Pixels

```python
#Use the range based indexing in python
test_roi=testImage[0:2,0:4]
print("Original Matrix\n{}\n".format(testImage))
print("Region of Interest\n{}\n".format(test_roi))
```

Modifying the ROI

```python
testImage[0:2,0:4]=110
print("Modified Matrix\n{}\n".format(testImage))
```

# Color Images

```python
# Path of the color image
imagePath="musk.jpg"

# Read the image
image=cv2.imread(imagePath)

# Print the image dimensions
print("Image Dimensions\n{}".format(image.shape))
```

This image is a color image, hence there are 3 channels (true for most of the color images)

Image Channels

```python
# Display Image
# Firt function: matplotlib imshow: first, convert BGR to RGB
# (because OpenCV use BGR format by default)
imgRGB=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB)

# or: reverse the order channels
# swap 1st and 3rd channel
#plt.imshow(image(:,:,::-1))
```

```python
# Second function
cv2.imshow("image",image)
cv2.waitKey()
cv2.destroyWindow("image")
```

```python
# Show the different channels
plt.figure(figsize=[20,5])

plt.subplot(131)
plt.imshow(image[:,:,0])
plt.title("Blue Channel")

plt.subplot(132)
plt.imshow(image[:,:,1])
plt.title("Green Channel")

plt.subplot(133)
plt.imshow(image[:,:,2])
plt.title("Red Channel")
```

#  Splitting and Merging Channels

```python
# Split the image in Blue, Green, and Red components
b,g,r=cv2.split(image)

# Show the different channels
plt.figure(figsize=[20,5])

plt.subplot(141); plt.imshow(image[:,:,0]); plt.title("Blue Channel")

plt.subplot(142); plt.imshow(image[:,:,1]); plt.title("Green Channel")

plt.subplot(143); plt.imshow(image[:,:,2]); plt.title("Red Channel")

# Merge the channels
imageMerged=cv2.merge((b,g,r))
# Show the merged image
plt.subplot(144);plt.imshow(imageMerged[:,:,::-1]);plt.title("Merged Output");

# or use the below functions
#cv2.imshow("mergedImage",imageMerged)
#cv2.waitKey()
#cv2.destroyWindow("mergedImage")
```

# Manipulating Color Pixels

```python
#Load and read the image
imagePath="number_zero.jpg"

#Read the image in Color format
testImage=cv2.imread(imagePath,1)
plt.imshow(testImage)
```

```python
# Modify pixel values
plt.figure(figsize=[20,20])

# Specify 3 values. Each for differnt channels
testImage[0,0]=(0,255,255)
plt.subplot(131)
plt.imshow(testImage[:,:,::-1])

testImage[1,1]=(255,0,255)
plt.subplot(132)
plt.imshow(testImage[:,:,::-1])
```

```python
# Modify ROI
testImage[0:3,0:3]=(255,0,0)
testImage[3:6,0:3] = (0,255,0)
plt.imshow(testImage[:,:,::-1])
```

# Alpha Channel Image

```python
imagePath="panther.png"

# Read the image
image=cv2.imread(imagePath,-1)
print("Image Dimension={}".format(image.shape))

# First three channels forms BGR image
imageBGR=image[:,:,0:3]

# Mask is the alpha channel of the original image
imageMask=image[:,:,3]

plt.figure(figsize=[15,15])
plt.subplot(121);plt.imshow(imageBGR[:,:,::-1]);plt.title('Color Image')
plt.subplot(122);plt.imshow(imageMask,cmap='gray');plt.title('Alpha Channel')
```
